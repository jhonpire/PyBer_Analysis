```python
%matplotlib inline
```

```python
# Import dependencies.
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot
plt.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt() Object-Oriented Interface
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt() Object-Oriented 2nd option
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt() Object-Oriented Quick option
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot. Matplotlib Graph with ptl.show
plt.plot(x_axis, y_axis)
plt.show()
```

```python
# Create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
# Create the plot.
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add a grid.
plt.grid()
# Add the legend.
plt.legend()
```

```python
# Create the plot with ax.plt() Object-Oriented Interface Skill-Drill
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis, marker='D', color='green', linewidth='2', label='Boston')
ax.legend()
ax.grid()
ax.set_title("Pyber Fare by Month")
ax.set_xlabel('Date')
ax.set_ylabel('Fare($)')
ax.set_ylim(0,45)
```

## Creating a Scatter Plot with the MATLAB method

```python
### There are two ways to create a scatter plot using the MATLAB approach
#### plt.plot()
#### plt.scatter()
```

```python
plt.plot(x_axis, y_axis, 'o')
```

```python
plt.scatter(y_axis, x_axis, color='red', label='Chicago')
#Add Legend for the city of Chicago
plt.legend()
#Create Labels for X and Y axis
plt.xlabel('Date')
plt.ylabel('Fare($)')
#Create a title
plt.title('PyBer Fare by Month')
#Invert axis
plt.gca().invert_yaxis()
#Add limit for the x-axis data
plt.xlim(0,50)
```

## Creating a Bubble Chart using the MATLAB method

```python
plt.scatter(x_axis, y_axis, s=y_axis)
```

```python
#Create a variable "Empty list" to hold the values of "y_axis" multiplied by 3
y_axis_larger = []
for data in y_axis:
  y_axis_larger.append(data*3)
plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

```python
plt.scatter(x_axis, y_axis, s = [i * 3 for i in y_axis])
```

## Creating a Bubble Chart using the Object-Oriented interface

```python
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis, s=y_axis)
```

```python
fig, ax = plt.subplots()
ax.scatter(y_axis, x_axis, s=[i * 5 for i in y_axis], label='Boston',
           alpha=0.8, color='skyblue', linewidths=2, edgecolors='black')
#Set label for Y Axis
ax.set_ylabel('Date')
#Set Label for X Axis
ax.set_xlabel('Fare ($)')
#Set a title for the plot
ax.set_title('Pyber Fare by Month')
#Set a Legend for the plot
ax.legend()
#Set a limit for the x-axis data
ax.set_xlim(0,50)
#Inver the y-axis
ax.invert_yaxis()
```

## Creating Piecharts using the MATLAB Method

```python
plt.pie(y_axis, labels=x_axis)
plt.show()
```

```python
#Increase the size of the Plot
plt.subplots(figsize=(8, 8))
#Set a tuple to contain the Exploding values
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
#Create a Pie Chart using plt.pie
plt.pie(y_axis, explode=explode_values, labels=x_axis, autopct='%.1f%%')
plt.show()
```

```python
# Assign 12 colors, one for each month.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8, 8))
plt.pie(y_axis,
    explode=explode_values,
    colors=colors,
    labels=x_axis,
    autopct='%.1f%%')

plt.show()
```

## Creating a Pie Chart using the Object-Oriented Interface

```python
fig, ax = plt.subplots(figsize=(9,9))
#Create a vaiable containing colors
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
#Define Explode Values
explode_values = (0, 0, 0.2, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
ax.pie(y_axis, colors=colors,
       labels=x_axis,
       shadow='true', 
       startangle=95, 
       autopct='%1.1f%%',
       explode=explode_values, counterclock='')
plt.show()
```
